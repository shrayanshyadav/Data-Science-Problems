```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from scipy.stats import ttest_1samp
```

```
cutlets_df=pd.read_csv("/content/drive/MyDrive/Asignments/Cutlets.csv")
cutlets_df
```

```
#H0-No significance difference in the diameter of Cutlets
#H1-Both the diameters are significant different different
#We will do the paired T test 
```

```
#we will find p value(probability value) .If P-value is greater than (or equal to) α, then Fail to reject the null hypothesis(Ho)i.e. You Accept your Null Hypothesis. And, if the P-value is less than α, then the you Reject your null hypothesis.
```

```
cutlets_df.describe()
```

```
cutlets_df[['Unit A','Unit B']].describe()
```

```
x = cutlets_df['Unit A']
y = cutlets_df['Unit A']
t, p_value  = stats.ttest_ind(x,y, axis = 0) 
print('P_Value:', p_value)
if p_value<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```
#Inputs rae 4 discrete variables, we can perform annova test but output is alo discrete , hence we need to perform Chi square test
```

```
#Ho= Proportions of Male and Female are same
#H1= Proportions of Male and Female are not same
```

```
Sales_df=pd.read_csv("/content/drive/MyDrive/Asignments/BuyerRatio.csv")
Sales_df
```

```
Sales_df.info()
```

```
from scipy.stats import chi2_contingency 
```

```
import scipy.stats as stats
```

```
Sales_df
```

```
Sales_df=Sales_df.drop(["Observed Values"],axis=1)
```

```
chi2_stas,p_value,dof,ex=stats.chi2_contingency(Sales_df)
print(chi2_stas,p_value,dof,ex)
```

```
print('P_Value:', p_value)
if p_value<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```
#Hence as p value>0.05 we wil not reject the null hypothesis
#hence proportion of male and female are same accross regions
```
